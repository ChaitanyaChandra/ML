```python
import numpy as np
import pandas as pd
```

```python
# read the dataset
dataset =  pd.read_csv("/kaggle/input/data-csv/Data.csv")
dataset
```

```python
# split dependent and independent data 
independent_x = dataset.iloc[:,:-1].values
dependent_y = dataset.iloc[:,-1].values
independent_x, dependent_y
```

```python
# fix missing values in independent data
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan, strategy="mean")
independent_x[:,1:] = si.fit_transform(independent_x[:,1:])
independent_x
```

```python
# encode categorical data in independent data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(
    transformers=[("encoder", OneHotEncoder(), [0])],
    remainder="passthrough"
)
independent_x = ct.fit_transform(independent_x)
independent_x
```

```python
# encode categorical data in dependent data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dependent_y = le.fit_transform(dependent_y)
dependent_y
```

```python
# split data to train and test 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent_x, dependent_y, test_size=0.2, random_state=1)
x_train, x_test, y_train, y_test
```

```python
# standard scalar of data.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
x_train, x_test
```
